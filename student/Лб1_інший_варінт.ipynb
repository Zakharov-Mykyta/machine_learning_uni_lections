```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine, load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

wine = load_wine()
breast_cancer = load_breast_cancer()

scaler = StandardScaler()
wine_data = scaler.fit_transform(wine.data)
breast_cancer_data = scaler.fit_transform(breast_cancer.data)

pca_wine = PCA(n_components=2)
pca_breast_cancer = PCA(n_components=2)

wine_pca = pca_wine.fit_transform(wine_data)
breast_cancer_pca = pca_breast_cancer.fit_transform(breast_cancer_data)

wine_df = pd.DataFrame(data=wine_pca, columns=['PC1', 'PC2'])
breast_cancer_df = pd.DataFrame(data=breast_cancer_pca, columns=['PC1', 'PC2'])

# Visualizing the results
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].scatter(wine_df['PC1'], wine_df['PC2'], c=wine.target)
axs[0].set_title('Wine Dataset')
axs[1].scatter(breast_cancer_df['PC1'], breast_cancer_df['PC2'], c=breast_cancer.target)
axs[1].set_title('Breast Cancer Dataset')
plt.show()
```
